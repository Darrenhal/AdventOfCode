```python
### Puzzle #2 ###

import re

fields = ["byr", "iyr", "eyr", "hgt", "hcl", "ecl", "pid"]
eye_colors = ["amb", "blu", "brn", "gry", "grn", "hzl", "oth"]
hair_colors = re.compile(r"[#(0-9a-f){6}]")
height_cm = re.compile(r"(hgt:)([0-9]{1,3})(cm)")
height_in = re.compile(r"(hgt:)([0-9]{1,3})(in)")

valid_passports = 0

for line in clean_data:
    valid = True
    for field in fields:
        if not(field in line):
            valid = False
            break;
        elif field == "byr":
            year = int(line[line.index("byr:") + 4 : line.index("byr:") + 8])
            if year < 1920 or year > 2002:
                print("byr: " + line)
                valid = False
                break
        elif field == "iyr":
            year = int(line[line.index("iyr:") + 4 : line.index("iyr:") + 8])
            if year < 2010 or year > 2020:
                print("iyr: " + line)
                valid = False
                break
        elif field == "eyr":
            year = int(line[line.index("eyr:") + 4 : line.index("eyr:") + 8])
            if year < 2020 or year > 2030:
                print("eyr: " + line)
                valid = False
                break
        elif field == "hgt":
            field_index = line.index("hgt:")
            if re.search(height_cm, line):
                result = re.search(height_cm, line)
                height = int(result.group(2))
                print("hgt cm: " + str(height) + " l: " + line)
                if height < 150 or height > 193:
                    valid = False
                    break
            elif re.search(height_in, line):
                result = re.search(height_in, line)
                height = int(result.group(2))
                print("hgt in: " + str(height) + " l: " + line)
                if height < 59 or height > 76:
                    valid = False
                    break
        elif field == "hcl":
            hcl = line[line.index("hcl") + 4 : line.index("hcl") + 10]
            if not(hair_colors.match(hcl)):
                print("hcl: " + line)
                valid = False
                break
        elif field == "ecl":
            ecl = line[line.index("ecl") + 4 : line.index("ecl") + 7]
            if not(ecl in eye_colors):
                print("ecl: " + line)
                valid = False
                break
        else:
            valid = True
    if valid:
        valid_passports += 1

valid_passports
```
